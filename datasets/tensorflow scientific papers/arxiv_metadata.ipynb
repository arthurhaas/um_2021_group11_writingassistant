```python
## Importing packages

import json
import re
import random
import pickle
import numpy as np
import pandas as pd
import time
from tqdm import tqdm
import os

data_path = '../' 
```

```python
## Getting categories and titles for papers

arxiv_metadata = {}

with open(os.path.join(data_path,'arxiv-metadata-oai-snapshot.json')) as arxiv:
    for element in tqdm(arxiv):
        a = json.loads(element)
        title = re.sub(r"\s+"," ", a['title'])
        id = re.sub('/','',a['id'])
        cat = a['categories']#re.search(r'[a-zA-Z\-]+',a['categories']).group(0)
        arxiv_metadata[id] = {'category': cat, 'title': title}
```

```python
## Function for getting category and title from arxiv metadata dataset

def ifkeyerrorthennull(key):
    try:
        return arxiv_metadata[key]
    except:
        return 'NA'
```

```python
# Creating dictionary of dictionaries for the 200k papers

article_category = {}

with open(os.path.join(data_path,'arxiv-dataset','train.txt'), 'r') as scipapers:
    for art in tqdm(scipapers):
        article = json.loads(art)['article_id']
        article_category[article] = ifkeyerrorthennull(article)

```

```python
## Writing to pickle

with open(os.path.join('output','arxiv_metadata.pkl'), 'wb') as open_file:
    pickle.dump(article_category, open_file)
```
